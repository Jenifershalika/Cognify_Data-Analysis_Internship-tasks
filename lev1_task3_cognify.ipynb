```python
import pandas as pd

df = pd.read_csv(r"C:\Users\HP\Downloads\Dataset .csv")

df.head()
```

```python
price_counts = df['Price range'].value_counts()
price_counts
```

```python
price_percentage = (price_counts / len(df)) * 100

pd.DataFrame({
    'Restaurant_Count': price_counts,
    'Percentage (%)': price_percentage.round(2)
})
```

```python
import matplotlib.pyplot as plt

plt.bar(price_counts.index, price_counts.values)
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.title("Price Range Distribution")
plt.show()
```
