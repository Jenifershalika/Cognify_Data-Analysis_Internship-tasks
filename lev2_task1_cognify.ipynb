```python
import pandas as pd

df = pd.read_csv("Dataset .csv")   # example: zomato.csv
df.head()
```

```python
df.columns
```

```python
# Check first few rows
df[['Restaurant Name', 'City', 'Cuisines', 'Aggregate rating']].head()

# Check for missing or zero ratings
df['Aggregate rating'].isnull().sum()
df[df['Aggregate rating'] == 0]
```

```python
# Convert to numeric (just in case) and fill missing values with mean
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')
df['Aggregate rating'].fillna(df['Aggregate rating'].mean(), inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['Aggregate rating'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot distribution
plt.figure(figsize=(10,5))
sns.histplot(df['Aggregate rating'], bins=10, kde=False, color='skyblue')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.show()

# Find most common rating range
most_common_rating = df['Aggregate rating'].mode()[0]
print("Most common rating:", most_common_rating)
```

```python
avg_votes = df['Votes'].mean()
print("Average number of votes per restaurant:", round(avg_votes,2))
```

```python
# Categorize ratings for better insights
def rating_category(r):
    if r >= 4.5:
        return 'Excellent'
    elif r >= 4.0:
        return 'Good'
    elif r >= 3.0:
        return 'Average'
    else:
        return 'Poor'

df['Rating Category'] = df['Aggregate rating'].apply(rating_category)
df['Rating Category'].value_counts().plot(kind='bar', color='purple')
plt.title('Restaurants by Rating Category')
plt.show()
```
